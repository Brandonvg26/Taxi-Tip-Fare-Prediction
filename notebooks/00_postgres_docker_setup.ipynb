# Start a Postgres container (recommend to open docker dstp to ui confirm)

```python
# terminal
docker run -d `
  --name taxi_pg `
  -e POSTGRES_USER=taxi `
  -e POSTGRES_PASSWORD=taxi `
  -e POSTGRES_DB=taxi_db `
  -p 5432:5432 `
  postgres:15
```

### Verify connection

```python
%pip install psycopg2
```

```python
import psycopg2


conn = psycopg2.connect(
    host='localhost', port=5432,
    dbname='taxi_db', user='taxi', password='taxi'
)
cur = conn.cursor()
cur.execute('SELECT version()')
print(cur.fetchone())   # should print Postgres 15.x
conn.close()
```

### Create raw schema

```python
conn = psycopg2.connect(host='localhost',port=5432,dbname='taxi_db',user='taxi',password='taxi')
conn.autocommit = True
cur = conn.cursor()


cur.execute('CREATE SCHEMA IF NOT EXISTS raw;')
cur.execute('CREATE SCHEMA IF NOT EXISTS staging;')
cur.execute('CREATE SCHEMA IF NOT EXISTS marts;')
conn.close()
```
